lets add round max upto 2 after decimal


```python
def mean_of_list(numbers):
    """This function returns the mean of list of numbers"""
    return round(sum(numbers)/len(numbers), 2)
```

```python
prices = [10,20,2.4,40,50]
mean_of_list(prices)
```
